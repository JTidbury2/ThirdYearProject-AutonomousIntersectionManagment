```python
def load_model_for_inference(filename, input_shape, config):
    # Dummy 'env' with only observation_space to initialize the model
    class DummyEnv:
        observation_space = input_shape

    env = DummyEnv()
    agent = DQNAgent(env, config)  # Initialize with dummy env
    checkpoint = torch.load(filename, map_location=torch.device('cpu'))  # Load on CPU for inference
    agent.value_net.load_state_dict(checkpoint['state_dict'])
    agent.value_net.eval()  # Set to evaluation mode
    return agent.value_net  # Return only the value_net for inference
```
